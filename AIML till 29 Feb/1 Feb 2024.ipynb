```python
x_gaussian = np.random.randn(1000) # returns 1000 random values; no range
sigma = 5
mean = 60

x_shifted_gaussian_1 = x_gaussian * sigma + mean
x_shifted_gaussian_2 = x_gaussian * 7 + 70
```

```python
x_shifted_gaussian_1 = x_shifted_gaussian_1.round()
x_shifted_gaussian_2 = x_shifted_gaussian_2.round()
```

```python
plt.hist(x_shifted_gaussian_1, alpha = 0.5)
plt.hist(x_shifted_gaussian_2, alpha = 0.5)
```
