# Histogram


## Shifted Gaussian
shifted gaussian= gaussian * sigma + mean
- sigma is standard deviation

```python
x_gaussian = np.random.random(100)
```

```python
sigma = 5
mean = 70
```

```python
x_shifted_gaussian = x_gaussian*sigma + mean
```

```python
x_gaussian = x_gaussian.round()
x_shifted_gaussian = x_shifted_gaussian.round()
```

```python
plt.hist(x_gaussian)
plt.hist(x_shifted_gaussian)
```
