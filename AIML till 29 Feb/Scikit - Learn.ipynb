```python
dataset = pd.DataFrame(iris.data, columns = iris.feature_names)  # we can also give individual names to the columns
```

```python
dataset
```

```python
dataset['target'] = iris.target
```

```python
dataset
```

```python
# give target names according to target
dataset['Flower_name'] = iris.target_names[iris.target]
```

```python
dataset
```

```python
dataset['NAME'] = dataset.target.apply(lambda x: iris.target_names[x]) #same as above command
```

```python
dataset
```

```python
dataset = dataset.drop(columns = ['NAME'])
```

```python
dataset
```

```python
setosa = dataset[dataset['Flower_name'] == 'setosa']
versicolor = dataset[dataset['Flower_name'] == 'versicolor']
virginica = dataset[dataset['Flower_name'] == 'virginica']
```

```python
setosa.head()
```

```python
versicolor.head()
```

```python
plt.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker = "*", label = 'Setosa', color = 'cyan')
plt.scatter(versicolor['sepal length (cm)'], versicolor['sepal width (cm)'], marker = "^", label = 'Versicolor', color = 'red')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.title('Flowers')
plt.show()
```
