```python
df = df.drop(columns = ['show_id', 'title', 'director', 'cast', 'description'])
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), format = "%B %d, %Y") #%B is for month, %d is for day, and %Y is for year
```

```python
df.head()
```

```python
df['month'] = df['date_added'].dt.month
```

```python
df['day'] = df['date_added'].dt.day
```

```python
df['year'] = df['date_added'].dt.year
```

```python
df.head()
```

```python
df.hist()
plt.tight_layout()
plt.show()
```
